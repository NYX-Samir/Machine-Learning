```python
import pandas as pd
df=pd.read_csv(r"C:\Users\evilk\Downloads\salaries.csv")
df.head(10)
```

```python
input=df.drop(['salary_more_then_100k'],axis=1)
input
```

```python
target=df['salary_more_then_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
input['company_n']=le_company.fit_transform(input['company'])
input['job_n']=le_company.fit_transform(input['job'])
input['degree_n']=le_company.fit_transform(input['degree'])
```

```python
input_n=input.drop(['company','job','degree'],axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
Model_decisionTree=DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_n,target,test_size=0.3)
```

```python
x_test
```

```python
Model_decisionTree.fit(x_train,y_train)
```

```python
Model_decisionTree.predict(x_test)
```

```python
Model_decisionTree.score(x_test,y_test)
```

```python
Model_decisionTree.score(x_train,y_train)
```

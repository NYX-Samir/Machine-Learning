```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
dir(iris)
```

```python
iris['feature_names']
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']=iris.target
df
```

```python
import matplotlib.pyplot as plt
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.title("SCATTER PLOT")
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],marker='*',color='r')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],marker='.',color='b')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],marker='+',color='y')
```

```python
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.title("SCATTER PLOT")
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],marker='*',color='r')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],marker='.',color='b')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],marker='+',color='y')
```

```python
df['flower_name']=df.target.apply(lambda x:iris.target_names[x])
```

```python
df
```

```python
x=df.drop(['target','flower_name'],axis=1)
x
```

```python
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
Model_svc=SVC()
```

```python
Model_svc.fit(xtrain,ytrain)
```

```python
Model_svc.predict(xtest)
```

```python
Model_svc.score(xtest,ytest)
```

1. Regularization (C)

```python
Model_C=SVC(C=1)
Model_C.fit(xtrain,ytrain)
Model_C.score(xtest,ytest)
```

```python
Model_C=SVC(C=10)
Model_C.fit(xtrain,ytrain)
Model_C.score(xtest,ytest)
```

2.Gamma

```python
Model_gamma=SVC(gamma=1)
Model_gamma.fit(xtrain,ytrain)
Model_gamma.score(xtest,ytest)
```

```python
Model_gamma=SVC(gamma=10)
Model_gamma.fit(xtrain,ytrain)
Model_gamma.score(xtest,ytest)
```

3.Kernel

```python
Model_kernel=SVC(kernel='linear')
Model_kernel.fit(xtrain,ytrain)
Model_kernel.score(xtest,ytest)
```

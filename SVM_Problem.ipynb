Train SVM classifier using sklearn digits dataset (i.e. from sklearn.datasets import load_digits) and then,

    Measure accuracy of your model using different kernels such as rbf and linear.

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
import matplotlib.pyplot as plt
plt.gray()
for i in range(3):
    plt.matshow(digits.images[i])
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(digits.data,columns=digits.feature_names)
df
```

```python
X=df
```

```python
y=digits.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(len(X_train))
print(len(X_test))
print(len(X))
print(len(X_train)+len(X_test))
```

```python
from sklearn.svm import SVC
model_svc=SVC()
```

```python
model_svc.fit(X_train,y_train)
```

```python
model_svc.predict(X_test)
```

```python
model_svc.score(X_test,y_test)
```

1.Regularization

```python
model_C=SVC(C=9)
model_C.fit(X_train,y_train)
model_C.score(X_test,y_test)
```

2.Gamma

```python
model_Gamma=SVC(gamma=10)
model_Gamma.fit(X_train,y_train)
model_Gamma.score(X_test,y_test)
```

3.Kernel


```python
model_Kernel=SVC(kernel='rbf')
model_Kernel.fit(X_train,y_train)
model_Kernel.score(X_test,y_test)
```

```python
model_Kernel=SVC(kernel='linear')
model_Kernel.fit(X_train,y_train)
model_Kernel.score(X_test,y_test)
```
